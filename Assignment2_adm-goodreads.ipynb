```python
def rec1score(u, i, userHistory):
    if len(userHistory) == 0:
        return 0
    averageSim = []
    s1 = usersPerBook[i].difference(set([u]))
    for h in userHistory:
        s2 = usersPerBook[h].difference(set([u]))
        averageSim.append(Jaccard(s1,s2))
    averageSim = sum(averageSim)/len(averageSim)
    return averageSim

# 2: Jaccard similarity with most similar user who has consumed i
def rec2score(u, i, userHistory):
    bestSim = None
    for v in usersPerBook[i]:
        if u == v:
            continue
        sim = Jaccard(userHistory, booksPerUser[v])
        if bestSim == None or sim > bestSim:
            bestSim = sim
    if bestSim == None:
        return 0
    return bestSim

# Generate a recommendation for a user based on a given scoring function
def rec(u, score):
    history = booksPerUser[u]
    if len(history) > 5: # If the history is too long, just take a sample
        history = random.sample(history,5)
    bestBook = None
    bestScore = None
    for i in books:
        if i in booksPerUser[u]: continue
        s = score(u, i, history)
        if bestBook == None or s > bestScore:
            bestBook = i
            bestScore = s

    return bestBook, bestScore
```

```python
def recTest(simFunction, nUserSamples):
    books = set(usersPerBook.keys())
    users = list(booksPerUser.keys())
    
    better = 0
    worse = 0

    for u in random.sample(users, nUserSamples):
        booksU = set(booksPerUser[u])
        if len(booksU) < 2:
            continue
        i = random.sample(booksU, 1)[0]
        uWithheld = booksU.difference(set([i]))
        j = random.sample(books,1)[0]

        si = simFunction(u,i,uWithheld)
        sj = simFunction(u,j,uWithheld)
        
        if si > sj:
            better += 1
        if sj > si:
            worse += 1
            
    print("Better than random " + str(better) + " times")
    print("Worse than random " + str(worse) + " times")
```
